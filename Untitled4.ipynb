```python
from ucimlrepo import fetch_ucirepo as fu
import pandas as pd

adult = fu(id = 2)

x = adult.data.features
y = adult.data.targets

tab = pd.concat([x, y], axis = 1)

adult.variables
```

```python
# 1)  Число столбцов в наборе данных
print("1: ", len(tab.columns))

# 2)  Есть ли пропуски в данных? Если есть, то в каких столбцах
print("2: ", f"yes: {tab.columns[tab.isna().any()].tolist()}"  if len(tab.columns[tab.isna().any()].tolist()) > 0 else "no")

# 3)  Кол-во уникальных значений в столбце race
print("3: ", len(tab['race'].unique()))

# 4)  Медиана hours-per-week
print("4: ", tab['hours-per-week'].median())

# 5)  Кого больше - женщин или мужчин с ЗП >50K?
print("5: ", "male" if
    len(tab[(tab['income'] == ">50K") & (tab['sex'] == "Male")]) >
    len(tab[(tab['income'] == ">50K") & (tab['sex'] == "Female")]) else
    "female")

# 6)  Заполните пропущенные данные в отдельных столбцах наиболее встречаемыми значениями.
#     Как еще можно было бы заполнить пропущенные данные?
#     Ответ: заполнить либо медианой, либо наименьшим, либо, самое очевидное - средним
tab = tab.fillna(tab.mode().iloc[0])
print("6: ", len(tab.columns[tab.isna().any()].tolist()))
```
